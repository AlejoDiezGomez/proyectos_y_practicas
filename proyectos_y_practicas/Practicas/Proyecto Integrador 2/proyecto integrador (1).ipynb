```python
import pandas as pd
import requests
from bs4 import BeautifulSoup 
from datetime import datetime 
url = 'https://cuspide.com/100-mas-vendidos/'
page = requests.get(url)
soup = BeautifulSoup ( page.content , 'html.parser')
url_1 = 'https://dolarhoy.com/cotizaciondolarblue'
page_1 =requests.get(url_1)
soup_1 =BeautifulSoup ( page_1.content , 'html.parser')
url_2 = 'https://dolarhoy.com/cotizaciondolaroficial'
page_2 =requests.get(url_2)
soup_2 =BeautifulSoup ( page_1.content , 'html.parser')

#------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Nombre libros 
name = soup.find_all('h3' , class_="name product-title woocommerce-loop-product__title")
titulo = []
count= 0
for i in name:
    if count < 100:
      titulo.append(i.text)
    else:
       break
    count += 1  
#----------------------------------------------------------------------------------------------------------------------------------------------------------------------
#uRL 
div_elements = soup.find_all( class_="box-image-top")  # Encuentra todos los elementos 'class box img' en el HTML
element_list = []
for div_element in div_elements:
    a_element = div_element.find('a')  # Encuentra el primer elemento 'a' dentro del div
    if a_element:
        href_value = a_element.get('href')  # Obtiene el valor del atributo 'href'
        element_list.append (href_value) 
#print(element_list)
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------------
# PRecio 
Precio = soup.find_all(class_="woocommerce-Price-amount amount")
precio_list=[]
for i in Precio:
   precio_list.append(i.text)   
   precio_lista= precio_list[1:]#lista coregida
   lista_corregida = [precio.replace("\xa0", "").replace(".", "").replace(",", ".").replace("$"," ") for precio in precio_lista]
Precio_Float = [] # Valor  en float 
for elemento in lista_corregida:
    Precio__ = float(elemento)
    Precio_Float.append (Precio__)
#print(Precio_Float)
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#PRECIO EN DOLAR OFICIAL 
enlace_usd_oficial = soup_2.find() 
Value_usd_list = []

#--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
# Buscar el enlace con href="/cotizaciondolarblue"
enlace_blue = soup_1.find(href="/cotizaciondolarblue")

# Crear una lista para almacenar los valores
Value_Blue_list = []

# Verificar si se encontrÃ³ el enlace
if enlace_blue:
    # Dentro del enlace, buscar elementos con class="venta"
    valores_venta = enlace_blue.find(class_="venta")
    
    # Recorrer los elementos encontrados
    for valor_venta in valores_venta:
        # Obtener el texto del elemento
        texto_valor_venta = valor_venta.get_text(strip=True)
        
        # Agregar el valor a la lista
        Value_Blue_list.append(texto_valor_venta)
Valor_Blue_ = [] # Valor Blue en float 
for elemento in Value_Blue_list:
    Valor_Blue = float(elemento)
    Valor_Blue_.append (Valor_Blue)
#print(Valor_Blue_)

# Imprimir la lista de valores
#print(Value_Blue_list)
#------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#TRANSFORMAR VALOR EN PESOS A DOLAR BLUE
Conversion_a_blue = []
for elementos in Precio_Float:
    Conversion_a_blue.append(round(elementos/Valor_Blue_[0],2))

#print(Conversion_a_blue)
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#COLOCAR LA FECHA 
now = datetime.now()
fecha_de_tabla = now.date()

data = {'Top_100_libros': titulo, 'Url' : element_list, 'Precio_en_ARS': Precio_Float, 'Precio en usd blue': Conversion_a_blue , 'Fecha_de_Tabla':fecha_de_tabla}
df= pd.DataFrame(data)
print(df)
df.to_csv('top100librosdelasemana.csv', index=False, sep=';', encoding='utf-8-sig')
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- 
#ACCESO A LA DATBASE DESDE PYTHON PARA SU MANIPULACION 
import pymysql
conexion = pymysql.connect (
    host = 'localhost',
    user= 'root',
    passwd= 'root1234',
    db = 'cuspidetop100',
)

#OBJETO CURSOR PARA LA EJECUCION DE LOS QUERY 

cursor = conexion.cursor()
 
insertQuery = "INSERT INTO top100libros (titulo , url , precio , precioUsd, precioUsdBlue , fecha) VALUES (%s, %s, %s , %s , %s , %s)"
datosInsertar = (titulo , element_list , Precio_Float , Conversion_a_blue , fecha_de_tabla )
```
