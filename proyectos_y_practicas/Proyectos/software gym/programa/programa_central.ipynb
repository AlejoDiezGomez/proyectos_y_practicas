```python
#funcion para registrar alumno y crear carnet
import pymysql
def ingresar_alumnos_y_crear_carnet():
 try:
        conexion = pymysql.connect(
        host= 'localhost',
        user='root',
        passwd='root1234',
        db='appgym3'
        )
        cursor = conexion.cursor()
        
        conexion.begin()

        nombre = input("Ingrese el nombre del alumno: ")
        apellido = input("Ingrese el apellido del alumno: ")
        telefono = input("Ingrese el número de teléfono del alumno: ")
        fecha_inscripcion = input("Ingrese la fecha de inscripción (AAAA-MM-DD): ")
        pago = input('ingrese monto pagado')
        fecha_inicio_carnet = input ('Fecha de pago (AAAA-MM-DD):')
        fecha_vencimiento = input ('fecha de vencimiento (AAAA-MM-DD):')

        insert_query = "INSERT INTO alumnos_registro (nombre, apellido, telefono, fecha_inscripcion) VALUES (%s ,%s , %s ,%s)"
        datos_a_insertar = (nombre, apellido, telefono, fecha_inscripcion)
        cursor.execute(insert_query , datos_a_insertar)
        conexion.commit()
        
        # Obtener el ID del último registro insertado en la tabla "alumnos_registro"
        cursor.execute("SELECT LAST_INSERT_ID()")
        ultimo_id = cursor.fetchone()[0]


        insert_query_carnet = "INSERT INTO carnet (IDregistro, fecha_inicio_carnet , fecha_vencimiento_carnet , pago) VALUES (%s ,%s , %s , %s)"
        datos_a_insertar_carnet = (ultimo_id, fecha_inicio_carnet, fecha_vencimiento, pago)
        cursor.execute(  insert_query_carnet ,  datos_a_insertar_carnet)
        conexion.commit()

        print('alumno y carnet registrados con exito')
 except pymysql.Error as e :  
        print(f"Error: {e}")
        # Si ocurre un error, deshacer la transacción
        conexion.rollback()
 finally:
    cursor.close()
    conexion.close()


ingresar_alumnos_y_crear_carnet()
```
